### 3.	Do customer segmentation appropriately. Please explain in detail and completely

## RFM

### Recency (R)
Berapa hari dari hari terakhir customer beli barang

```python
datar['Date'] = pd.DatetimeIndex(datar['Tanggal_Transaksi']).date
```

```python
#Tanggal Transaksi paling terakhir
datar['Tanggal_Transaksi'].max()
now = dt.date(2011,12,9)
print(now)
```

```python
recency_df = datar.groupby(by='Kode_Pelanggan', as_index=False)['Date'].max()
recency_df.columns = ['Kode_Pelanggan','Tanggal_Transaksi_Terakhir']
recency_df.head()
```

```python
recency_df['Recency'] = recency_df['Tanggal_Transaksi_Terakhir'].apply(lambda x: (now - x).days)
recency_df.head()
```

### Frequency (F)
Berapa banyak inovoice per customer

```python
# Drop nilai duplicate
datar_copy = datar
datar_copy.drop_duplicates(subset=['Kode_Bayar', 'Kode_Pelanggan'], keep="first", inplace=True)
```

```python
frequency_df = datar_copy.groupby(by=['Kode_Pelanggan'], as_index=False)['Kode_Bayar'].count()
frequency_df.columns = ['Kode_Pelanggan','Frequency']
frequency_df.head()
```

### Monetary
Berapa yang customer bayar ke kita

```python
datar['Total'] = datar['Harga_Satuan'] * datar['Jumlah']
```

```python
monetary_df = datar.groupby(by='Kode_Pelanggan',as_index=False).agg('sum')
monetary_df.head()
```

### Create RFM Table

```python
temp_df = recency_df.merge(frequency_df,on='Kode_Pelanggan')
temp_df.head()
```

```python
rfm_df = temp_df.merge(monetary_df,on='Kode_Pelanggan')
```

```python
#use CustomerID as index
rfm_df.set_index('Kode_Pelanggan',inplace=True)
#check the head
rfm_df.head()
```

### RFM Model

```python
quantiles = rfm_df.quantile(q=[0.25,0.5,0.75])
quantiles
quantiles.to_dict()
```

```python
# x = value, rfm = recency, monetary_value, frequency, q = quartiles qict
def RScore(x,rfm,q):
    if x <= q[rfm][0.25]:
        return 4
    elif x <= q[rfm][0.50]:
        return 3
    elif x <= q[rfm][0.75]: 
        return 2
    else:
        return 1
# x = value, rfm = recency, monetary_value, frequency, k = quartiles qict
def FMScore(x,rfm,q):
    if x <= q[rfm][0.25]:
        return 1
    elif x <= q[rfm][0.50]:
        return 2
    elif x <= q[rfm][0.75]: 
        return 3
    else:
        return 4
```

```python
#RFM Segmentation Table
rfm_segmentation = rfm_df
rfm_segmentation['R_Quartile'] = rfm_segmentation['Recency'].apply(RScore, args=('Recency',quantiles,))
rfm_segmentation['F_Quartile'] = rfm_segmentation['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
rfm_segmentation['M_Quartile'] = rfm_segmentation['Total'].apply(FMScore, args=('Total',quantiles,))
```

```python
rfm_segmentation.head()
```

```python
rfm_segmentation['RFMScore'] = rfm_segmentation.R_Quartile.map(str) \
                            + rfm_segmentation.F_Quartile.map(str) \
                            + rfm_segmentation.M_Quartile.map(str)
rfm_segmentation.head()
```
